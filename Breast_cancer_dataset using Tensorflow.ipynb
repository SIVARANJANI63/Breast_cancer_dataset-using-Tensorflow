```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()#relational data like data target columns name
```

```python
breast_cancer_dataset
```

```python
import pandas as pd
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
data_frame['label']=breast_cancer_dataset.target
```

```python
x=data_frame.drop('label',axis=1)
y=data_frame['label']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
for i in x.columns:
    x[i]=sc.fit_transform(x[[i]])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')
                        ])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,validation_split=0.1,epochs=50)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
y_pred=model.predict(x_test)
```
